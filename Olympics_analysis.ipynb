```python
def fetch_medal_tally(year, country):
    medal_tally = df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])

    if year == 'Overall' and country == 'Overall':
        temp_df = medal_tally
    elif year == 'Overall' and country != 'Overall':
        temp_df = medal_tally[medal_tally['region'] == country]
    elif year != 'Overall' and country == 'Overall':
        temp_df = medal_tally[medal_tally['Year'] == year]
    else:
        temp_df = medal_tally[(medal_tally['Year'] == year) & (medal_tally['region'] == country)]

    if temp_df.empty:
        return pd.DataFrame(columns=['Gold','Silver','Bronze','total'])   # return empty result instead of error

    if year == 'Overall':
        x = temp_df.groupby('NOC')['Medal'].value_counts().unstack(fill_value=0)
    else:
        x = temp_df.groupby('Year')['Medal'].value_counts().unstack(fill_value=0)

    # Ensure missing medal types don’t cause KeyError
    x = x.reindex(columns=['Gold','Silver','Bronze'], fill_value=0)

    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']
    return x
```

```python
def fetch_medal_tally(year, country):
    # Step 1: remove duplicate medals (team events double count na ho)
    medal_tally = df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])

    flag = 0   # 👈 flag initialize

    # Step 2: filter year and country
    if year == 'Overall' and country == 'Overall':
        temp_df = medal_tally

    elif year == 'Overall' and country != 'Overall':
        flag = 1
        temp_df = medal_tally[medal_tally['region'] == country]

    elif year != 'Overall' and country == 'Overall':
        temp_df = medal_tally[medal_tally['Year'] == year]

    else:
        flag = 1
        temp_df = medal_tally[(medal_tally['Year'] == year) & (medal_tally['region'] == country)]

    # 🔹 Handle empty dataset
    if temp_df.empty:
        return pd.DataFrame(columns=['Gold', 'Silver', 'Bronze', 'total'])

    # Step 3: groupby (flag decide karega grouping kis column pe hogi)
    if flag == 1:
        x = temp_df.groupby('NOC')['Medal'].value_counts().sort_values().unstack(fill_value=0)
        #x = temp_df.groupby('Year').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Year').reset_index()
    else:
        x = temp_df.groupby('Year')['Medal'].value_counts().sort_values().unstack(fill_value=0)
        #x = temp_df.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()

    # Step 4: total column
    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']

    return x
```
